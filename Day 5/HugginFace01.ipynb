```python
from transformers import pipeline
classifier=pipeline('sentiment-analysis')
```

Pre-trained model get downloaded

```python
classifier("The product doesn't perform good.")
```

The model can be used on the list of sentenses, which will be preprocessed then fed ot the model as a batch, returning a list of dictionaries.

```python
result=classifier(["The product really peroforms very well", "Movie was incrediable"])
for res in result:
  print(f"label: {res['label']},  score: {round(res['score'],4)}")
```

AutoTokenizer: Text data into numerical data

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
# This model only exists in PyTorch, so we use the `from_pt` flag to import that model in TensorFlow.
model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
```

```python
classifier("Food was realy great")
```

```python
inputs = tokenizer("Experience was really too cool")

print(inputs)
```

You can pass a list of sentences directly to your tokenizer. If your goal is to send them through your model as a
batch, you probably want to pad them all to the same length, truncate them to the maximum length the model can accept
and get tensors back. You can specify all of that to the tokenizer:

```python
tf_batch = tokenizer(
    ["Experience was really too cool", "Food was not good"],
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="tf"
)
```
